```python
import pandas as pd
```

```python
tree_census = pd.read_csv('data.csv')
tree_census.head(3)
```

```python
tree_census.columns
```

```python
tree_census_subset = tree_census[['tree_dbh', 'stump_diam',
                                  'curb_loc', 'status', 'health', 'spc_latin', 'spc_common', 'steward',
                                  'sidewalk', 'user_type', 'problems', 'root_stone',
                                  'root_grate', 'root_other', 'trunk_wire', 'trnk_light', 'trnk_other',
                                  'latitude', 'longitude',
                                  'brch_light', 'brch_shoe', 'brch_other', ]]
tree_census_subset
```

```python
print(tree_census_subset.dtypes)
```

```python
tree_census_subset.isna().sum()
```

```python
tree_census_subset[tree_census_subset['health'].isna()]
```

```python
tree_census_subset['tree_dbh'].fillna(tree_census_subset['tree_dbh'].mean(), inplace=True)
tree_census_subset['stump_diam'].fillna(tree_census_subset['stump_diam'].mean(), inplace=True)
```

```python
tree_census_subset['steward'].fillna('Unknown', inplace=True)
tree_census_subset['sidewalk'].fillna('Unknown', inplace=True)
tree_census_subset['problems'].fillna('Unknown', inplace=True)
tree_census_subset['spc_latin'].fillna('Unknown', inplace=True)
tree_census_subset['spc_common'].fillna('Unknown', inplace=True)
```

```python
tree_census_subset.dropna(subset=['health'], inplace=True)
```

```python
print(tree_census_subset.isna().sum())
```

```python
from sklearn.preprocessing import LabelEncoder

cat_columns = ['curb_loc', 'status', 'health', 'spc_latin', 'spc_common', 'steward',
               'sidewalk', 'problems', 'root_stone', 'user_type',
               'root_grate', 'root_other', 'trunk_wire', 'trnk_light', 'trnk_other',
               'brch_light', 'brch_shoe', 'brch_other']

label_encoders = {}
for col in cat_columns:
    le = LabelEncoder()
    tree_census_subset[col] = le.fit_transform(tree_census_subset[col])
    label_encoders[col] = le

print(tree_census_subset.head())
```
